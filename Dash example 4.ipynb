```python
import dash
from dash.dependencies import Output, Input
import dash_core_components as dcc
import dash_html_components as html
import plotly
import random
import plotly.graph_objs as go
from collections import deque

# setting initial values and maximum number of plots
X = deque(maxlen = 20)
Y = deque(maxlen = 20)
X.append(1)
Y.append(1)
```

```python
app = dash.Dash(__name__)

#configuring live graph, and interval length
app.layout = html.Div([
    dcc.Graph(id = 'live-graph', animate = True),
    dcc.Interval (
        id = 'graph-update',
        interval = 1000,
        n_intervals = 0)
])


@app.callback(Output('live-graph', 'figure'),
              [Input('graph-update', 'n_intervals')])


def update_graph(n):
    ## function to update data, an SQL can go here
    X.append(X[-1]+1)
    Y.append(Y[-1]+(Y[-1]*random.uniform(-0.1,0.1)))
    
    #type of graph
    data = go.Scatter(
        x = list(X),
        y = list(Y),
        name = 'Scatter',
        mode = 'lines+markers'
        )
    ##return graph with changing ranges to coincide with max number of plots
    return {'data':[data], 'layout': go.Layout(xaxis = dict(range = [min(X), max(X)]),
                                             yaxis = dict(range = [min(Y), max(Y)]))
           }


 #running server               
if __name__ == '__main__':
    app.run_server()                    
```
